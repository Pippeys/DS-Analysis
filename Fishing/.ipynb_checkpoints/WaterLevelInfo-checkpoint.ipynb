```python
import hydrofunctions as hf
import pandas as pd
myTime = 'P20D'
main_df = hf.NWIS('11161000','dv',period=myTime)
name = 'San_Lorenzo_River'
main_df = main_df.get_data().df()
main_df.rename(columns={main_df.columns[0]:name},inplace=True)
main_df.drop(columns=main_df.columns[1],inplace=True)
library = {'Upper_San_Lorenzo_River':'11160500',
           'Guerneville_Russian_River':'11467000',
           'Putah_Creek':'11454000',
           'Farad_Truckee_River':'10346000',
           'Little_Truckee':'10344400'
          }
```

```python
for name,river in library.items():
    river_data = hf.NWIS(river,'dv',period=myTime)
    df = river_data.get_data().df()
    df.rename(columns={df.columns[0]:name},inplace=True)
    df.drop(columns=df.columns[1],inplace=True)
    main_df = main_df.join(df)
main_df.tail()
```

```python
from climata.acis import StationDataIO
import time
import datetime as dt
big = pd.DataFrame()
parameters = ['pcpn','avgt']
basins_library = {'TRUCKEE #2':'16050102','GRATON':'18010110'}
start = dt.datetime.now() - dt.timedelta(days=20)
end = dt.datetime.now()

d = {}
for name in basins_library
    d[name] = pd.DataFrame()

x=0
i = 0

while i < 2:
    while i == 0:
        para = parameters[i]
        for basin,station in basins_library.items():
                sites = StationDataIO(
                    basin = station,
                    start_date = start,
                    end_date = end,
                    parameter = para)
                for site in sites:
                    if site[0] == basin:
                        if para == 'pcpn':
                            for evt in site.data:
                                d = d.append({'Date':evt.date,basin[0:5]+'_pcpn':evt.pcpn},ignore_index=True)                 
            x=x+1
        i = i+1
    while i == 1:
        a = pd.DataFrame()
        para = parameters[i]
        for basin,station in basins_library.items():
            sites = StationDataIO(
                basin = station,
                start_date = start,
                end_date = end,
                parameter = para)
            for site in sites:
                if site[0] == basin:
                    if para == 'avgt':
                        for evt in site.data:
                            a = a.append({basin[0:5]+'_avgt':evt.avgt},ignore_index=True)
        i = i+1
    x = pd.concat([a,d],axis=1)
    big = pd.concat([x,big],axis=1)
```

```python
main_df.join(big.set_index(['Date']))
```
