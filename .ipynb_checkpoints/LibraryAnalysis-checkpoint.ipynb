## Library Usage

Analyze for different groups of people based their library activities.

```python
#import and read in data

import pandas as pd 
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline
sb.set_style('whitegrid')
fig = plt.figure()

data = pd.read_csv('https://data.sfgov.org/api/views/qzz6-2jup/rows.csv?accessType=DOWNLOAD')
```

Assumptions: 
- There are no incorrect values
- No need to standardize continous attributes

```python
#Briefly look at data
df = data
df.head()
```

### Duplicates Columns
- Patron Type Code & Patron Type Definition
- Home Library Code & Home Library Definition
- Notice Preference Code & Notice Preference Definition

```python
df = df.drop('Patron Type Code',1)
df = df.drop('Home Library Code',1)
df = df.drop('Notice Preference Code',1)
```

### Remove Nulls

```python
print(df.isnull().sum())
df = df.drop('Supervisor District',1)
df = df.dropna(axis=0)
```

### Explore!

```python
#
## First let's see the spread of checkouts and renewals (activities)
## Since there are 420k records lets take a small sample of the total population...let's say 5%
#
df_sample = df.sample(frac=.05)
sb.regplot(df_sample['Total Checkouts'],df_sample['Total Renewals'], fit_reg=False)
```

Normally I would standardize these 2 variables because they are really really skewed. But I want to look at other attributes more.

### Let's look at all the defining attributes of the population's library activity 
1. Patron Type
2. Age Range
3. Library Location
4. Last Active Year
5. Preferred Notification medium

```python

# Patron Type
grouping = df.groupby('Patron Type Definition')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Patron Type by Number of Activities')
plt.xticks(rotation= 90)
plt.show()

# Age Range
grouping = df.groupby('Age Range')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Age Group by Number of Activities')
plt.xticks(rotation= 90)
plt.show()

# Library Locaiton
grouping = df.groupby('Home Library Definition')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Library Branch Activity')
plt.xticks(rotation= 90)
plt.show()

# Last Active year
grouping = df.groupby('Circulation Active Year')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Last Active Year')
plt.xticks(rotation= 90)
plt.show()

# Preferred notification medium
grouping = df.groupby('Notice Preference Definition')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Preferred notification medium')
plt.xticks(rotation= 90)
```

Alright! Sweet, so we have all the important attributes all spread out to look at.



Let's slice these to see what are some groups!
    - lets have 3 groups to look at:
        - High Checkouts & High Renewals
        - Low Checkouts & Low Renewals
        - Outliers of Both
    - High Checkout/Renewals will be > median
    - Low Checkout/Renewals will be < median
    - Outliers will be > quantile 3


## High Checkouts/Renewals Slice

```python
q3c = df['Total Checkouts'].quantile(.9)
q3r = df['Total Renewals'].quantile(.9)
print(q3c)
print(q3r)
med = np.mean(df['Total Checkouts'])
print(med)
mid = np.mean(df['Total Renewals'])
print(mid)
df_slice = df.loc[df['Total Checkouts'] > med]
df_slice = df_slice.loc[df_slice['Total Checkouts'] < q3c]
df_slice = df_slice.loc[df_slice['Total Renewals'] > mid]
df_slice = df_slice.loc[df_slice['Total Renewals'] < q3r]
print(len(df_slice))
df_sample = df_slice
sb.regplot(df_sample['Total Checkouts'],df_sample['Total Renewals'], fit_reg=False)
plt.show()

# Patron Type
grouping = df_slice.groupby('Patron Type Definition')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Patron Type by Number of High Activities')
plt.xticks(rotation= 90)
plt.show()

# Age Range
grouping = df_slice.groupby('Age Range')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Age Group by Number of High Activities')
plt.xticks(rotation= 90)
plt.show()

# Library Locaiton
grouping = df_slice.groupby('Home Library Definition')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Library Branch by High Activities')
plt.xticks(rotation= 90)
plt.show()

# Last Active year
grouping = df_slice.groupby('Circulation Active Year')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Last Active Year by High Activities')
plt.xticks(rotation= 90)
plt.show()

# Preferred notification medium
grouping = df_slice.groupby('Notice Preference Definition')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Preferred notification medium by High Activities')
plt.xticks(rotation= 90)
```

## Low Checkouts/Renewals Slice

```python
med = np.mean(df['Total Checkouts'])
print(med)
mid = np.mean(df['Total Renewals'])
print(mid)
df_slice = df.loc[df['Total Checkouts'] < med]
df_slice = df_slice.loc[df_slice['Total Renewals'] < mid]
print(len(df_slice))
df_sample = df_slice
sb.regplot(df_sample['Total Checkouts'],df_sample['Total Renewals'], fit_reg=False)
plt.show()

# Patron Type
grouping = df_slice.groupby('Patron Type Definition')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Patron Type by Number of Low Activities')
plt.xticks(rotation= 90)
plt.show()

# Age Range
grouping = df_slice.groupby('Age Range')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Age Group by Number of Activities')
plt.xticks(rotation= 90)
plt.show()

# Library Locaiton
grouping = df_slice.groupby('Home Library Definition')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Library Branch Low Activities')
plt.xticks(rotation= 90)
plt.show()

# Last Active year
grouping = df_slice.groupby('Circulation Active Year')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Last Active Year by Low Activities')
plt.xticks(rotation= 90)
plt.show()

# Preferred notification medium
grouping = df_slice.groupby('Notice Preference Definition')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Preferred notification medium by Low Activities')
plt.xticks(rotation= 90)
```

## Activity Outliers Slice

```python
q3c = df['Total Checkouts'].quantile(.9)
q3r = df['Total Renewals'].quantile(.9)
print(q3c)
print(q3r)
df_slice = df.loc[df['Total Checkouts'] > q3c]
df_slice = df_slice.loc[df_slice['Total Renewals'] > q3r]

df_sample = df_slice
sb.regplot(df_sample['Total Checkouts'],df_sample['Total Renewals'], fit_reg=False)
plt.show()

# Patron Type
grouping = df_slice.groupby('Patron Type Definition')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Patron Type by Number of Outlying Activities')
plt.xticks(rotation= 90)
plt.show()

# Age Range
grouping = df_slice.groupby('Age Range')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Age Group by Number of Outlying Activities')
plt.xticks(rotation= 90)
plt.show()

# Library Locaiton
grouping = df_slice.groupby('Home Library Definition')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Library Branch of Outlying Activities')
plt.xticks(rotation= 90)
plt.show()

# Last Active year
grouping = df_slice.groupby('Circulation Active Year')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Last Active Year of Outlying Activities')
plt.xticks(rotation= 90)
plt.show()

# Preferred notification medium
grouping = df_slice.groupby('Notice Preference Definition')
num = grouping['Outside of County'].count().sort_values(ascending=False)
ax = sb.barplot(x=num.index,y=num)
plt.ylabel('Count')
plt.title('Preferred notification medium of Outlying Activities')
plt.xticks(rotation= 90)
```

### Quick run of KMeans Cluster and RandomForest Classifying

#### KMeans Clustering

```python
from sklearn.cluster import KMeans
from scipy import stats
from sklearn.preprocessing import scale 
from sklearn.preprocessing import LabelEncoder

meaner = df[['Total Checkouts','Total Renewals']]
num = LabelEncoder()
meaner['bin_age'] = num.fit_transform(df['Age Range'].astype('str'))
meaner['bin_Library'] = num.fit_transform(df['Home Library Definition'].astype('str'))
meaner['bin_Month'] = num.fit_transform(df['Circulation Active Month'].astype('str'))
meaner['bin_year'] = num.fit_transform(df['Circulation Active Year'].astype('str'))
meaner['bin_notice'] = num.fit_transform(df['Notice Preference Definition'].astype('str'))
meaner['bin_email'] = num.fit_transform(df['Provided Email Address'].astype('str'))
meaner['bin_registered'] = num.fit_transform(df['Year Patron Registered'].astype('str'))
meaner['bin_county'] = num.fit_transform(df['Outside of County'].astype('str'))
meaner = meaner.sample(frac=.05)
kmeans = KMeans(n_clusters=5)
kmeans = kmeans.fit(meaner)
labels = kmeans.labels_


plt.figure(figsize=(8,6))
plt.scatter(meaner['Total Checkouts'], meaner['Total Renewals'], c=labels,cmap='rainbow')
plt.ylabel('Total Renewals')
plt.xlabel('Total Checkouts')
plt.show()
```

This graphic hints at an idea of how a classifier may work.

#### RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

data = df.sample(frac=.05)
datay = data['Age Range']

data = data.drop('Patron Type Definition',1)
data = data.drop('Total Checkouts',1)
data = data.drop('Total Renewals',1)

datax = pd.get_dummies(data)
y = datay
x = datax.join(df['Total Checkouts'])
x = x.join(df['Total Renewals'])     
                              
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)

classifier = RandomForestClassifier(max_depth=2, random_state=0)
classifier.fit(x_train,y_train)
predict = classifier.predict(x_test)

print('train')
print(classifier.score(x_train,y_train))
print('test')
print(classifier.score(x_test,y_test))

true_rfc = x_test[['Total Checkouts','Total Renewals']]
true_rfc['true_type'] = y_test
sb.pairplot(true_rfc, hue = 'true_type', palette = 'hls')

pred_rfc = x_test[['Total Checkouts','Total Renewals']]
pred_rfc['predicted_range'] = predict
sb.pairplot(pred_rfc, hue = 'predicted_range', palette = 'hls')
```

These are the predicted groupings of train and test models. With the score being around 80% it seems there may be something worth diving into here.
